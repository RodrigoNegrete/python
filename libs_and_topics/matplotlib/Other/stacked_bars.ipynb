```python
# https://www.geeksforgeeks.org/create-a-stacked-bar-plot-in-matplotlib/

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.ticker as ticker
  
```

```python
# create data
df = pd.DataFrame([
    ['A', 10, 20, 10, 26], 
    ['B', 20, 25, 15, 21], 
    ['C', 12, 15, 19, 6],
    ['D', 10, 18, 11, 19]],
        columns=['Team', 'Round 1', 'Round 2', 'Round 3', 'Round 4'])

# view data
print(df)
 
# plot data in stack manner of bar type
df.plot(x='Team', kind='bar', stacked=True, title='Stacked Bar Graph by dataframe')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Create data 
df = pd.DataFrame([['A', 10, 20, 10, 26]], columns=['Team', 'Round 1', 'Round 2', 'Round 3', 'Round 4'])

# Set the first column as the index
df.set_index('Team', inplace=True)

# View data
print(df)

# Function to format y-axis as currency
def currency(x, pos):
    return '${:,.0f}'.format(x)

# Plot data in stack manner of bar type
ax = df.plot(kind='bar', stacked=True, title='Stacked Bar Graph by dataframe')

# Set x and y labels
plt.xlabel("Escenario")
plt.ylabel("Monto")

# Apply currency formatting to the y-axis
ax.yaxis.set_major_formatter(FuncFormatter(currency))

# Ensure x-tick labels (like 'A') are displayed horizontally
plt.xticks(rotation=0)

# Show the plot
plt.show()
```

```python
# create data
x = ["Composición del Pago"]
y1 = np.array([50000])
y2 = np.array([15000])
y3 = np.array([5000])
y4 = np.array([7500])
y5 = np.array([3000])
y6 = np.array([8000])

# plot bars in stack manner
plt.bar(x, y1, color='#884ea0')
plt.bar(x, y2, bottom=y1, color='#7fb3d5')
plt.bar(x, y3, bottom=y1+y2, color='#a3e4d7')
plt.bar(x, y4, bottom=y1+y2+y3, color='#a9dfbf')
plt.bar(x, y4, bottom=y1+y2+y3+y4, color='#f9e79f')
plt.bar(x, y4, bottom=y1+y2+y3+y4+y5, color='#808b96')
plt.xlabel("Escenario")
plt.ylabel("Monto")
plt.legend(["Pago", "IMSS", "ICRV", "ISR", "ISN", "Admon"])
plt.title("Escenario de Nómina")
plt.show()
```
