### Example - gamefaq - top 10 games

```python
# get new response from new constructed url
response = requests.get('https://gamefaqs.gamespot.com/pc', headers = user_headers)
response.status_code
```

```python
# make new soup
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
# check text
# strip text
soup.get_text(strip = True)
```

```python
# Top 10 Games
soup.find_all(string="Top 10 Games")
```

```python
# view the other slices to prove we got what we need
soup.find_all(string = 'Top 10 Games')[0].find_parents()[0]
```

```python
# second instance
soup.find_all(string = 'Top 10 Games')[0].find_parents()[1]
```

```python
# third instance as above
soup.find_all(string = 'Top 10 Games')[0].find_parents()[2]
```

```python
# li is a list insode an ordered list ol
# inside li there's an href with the link
# inside li there's a div with the title
# instances in the soup above are separated by commas
# it's the third instance in the list what we look for - we use 2
# slice a parent to see content
soup.find_all(string = 'Top 10 Games')[0].find_parents()[2]
```

```python
# store instance in variable
top_ten_vita_games = soup.find_all(string = 'Top 10 Games')[0].find_parents()[2]
top_ten_vita_games
```

```python
# verify the element a tag
type(top_ten_vita_games)
```

```python
# find ol elements and find li elements inside
# li is usually inside ol
# find fnds the first one so we dont need to slice
for item in top_ten_vita_games.find('ol').find_all('li'):
    # print stripped text per iteration
    print(item.get_text())
```

```python
# find within a to see if result improves
for item in top_ten_vita_games.find('ol').find_all('li'):
    # print stripped text per iteration
    print(item.find('a').get_text())
```

```python
# strip to see if result improves
for item in top_ten_vita_games.find('ol').find_all('li'):
    # print stripped text per iteration
    print(item.find('a').get_text().strip())
```

```python
# strip should remove staces to left and right
# the above means that the rank and the name are in the same >content<
# slice to review the first the second instance
top_ten_vita_games.find('ol').find_all('li')[1].get_text()
```

```python
# use strip to remove new lines
top_ten_vita_games.find('ol').find_all('li')[1].get_text().strip()
```

```python
# string formatting
for item in top_ten_vita_games.find('ol').find_all('li'):
    # split where ever ther's a space
    print(item.find('a').get_text().strip().split())
```

```python
# string formatting
# this code will change all the spaces for one
for item in top_ten_vita_games.find('ol').find_all('li'):
    # join where ever ther's a space
    # 'text for joining'.join(item)
    print(' '.join(item.find('a').get_text().strip().split()))
```

## example 3 - gamefaq Upcoming releases 
- Japan 
- Stop iterating when you find some text

```python
# find PC's upcoming releases
# get response
import requests
from bs4 import BeautifulSoup

user_headers = {'User-Agent': 'Mozilla/5.0'}

url = 'https://gamefaqs.gamespot.com/pc'
response = requests.get(url, headers = user_headers)

response.status_code

# make soup
soup = BeautifulSoup(response.content, 'html.parser')

soup.find(string = 'Upcoming PC Game Releases')
```

```python
# find parents
soup.find(string = 'Upcoming PC Game Releases').find_parents()
```

```python
# get number of instances
len(soup.find(string = 'Upcoming PC Game Releases').find_parents())
```

```python
# improve readability
for item in range(11):
    print(soup.find(string = 'Upcoming PC Game Releases').find_parents())
    # print new line and asteriscs after each instance
    print('\n*****************************************')
```

```python
# for some rason the japan data is in !DOCTYPE portion
# get america first that's more clean
# the third parent seems to have the game info
soup.find(string = 'Upcoming PC Game Releases').find_parents()[2]
```

```python
# save to variable
north_america_games = soup.find(string = 'Upcoming PC Game Releases').find_parents()[2]
type(north_america_games)
```

```python
len(north_america_games)
```

```python
# iterate over 'dl' tags on north_america_games
for item in north_america_games.find_all('dl'):
    # print dt inside dl
    print(item.find('dt').get_text())
    # Iterate over dd inside dt
    for games in (item.find_all('dd')):
        # print text
        print(games.get_text())
```

```python
# the last element belongs to japan
# remove it by slicing north_america_games.find_all('dl')
for item in north_america_games.find_all('dl')[:-1]:
    # print dt inside dl
    print(item.find('dt').get_text())
    # Iterate over dd inside dt
    for games in (item.find_all('dd')):
        # print text
        print(games.get_text())
```
