```python
import requests
from bs4 import BeautifulSoup
import requests
import re

for page in range(100):
    page_number = page
    
    url = "http://sitl.diputados.gob.mx/LXIV_leg/curricula.php?dipt={page_number}".format(page_number=page_number)

    response = requests.get(url, headers = {'User-Agent': 'Mozilla/5.0'})
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'lxml')
        
        nombre = soup.find('td', class_ = "textocurrienc")
        if nombre != "":
            print(nombre.text)
            
```
